# Module 0: Installation

If you wants to run the code on your own computer (instead of using colab), please follow these steps

1. Download the notebooks to your own computer, say the path you download to is `YOUR-PATH`
1. Install miniconda on your computer according to this instructions: https://docs.conda.io/en/latest/miniconda.html 
1. Open terminal and run the following commands:
    - Make a new environment with python 3.8 by running this command: `conda create --name py38 python=3.8`
    - Activate the environment: `conda activate py38`
    - Install all required python packages: `pip install jupyterlab tensorflow scikit-learn numpy pandas plotly-express matplotlib connectionist`
    - Run the Jupyter Lab server: `jupyter lab --notebook-dir="YOUR-PATH"`
    - Note: **Do not close the terminal windows**
1. Once you had jupyter lab server running, you should see jupyter lab in a browser. You can double-click to open a file.

## Reopening jupyter lab

Next time you want to open jupyter lab, you only need to do these in the terminal:
1. `conda activate py38`
2. `jupyter lab --notebook-dir="YOUR-PATH"`
